```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
df = pd.read_csv('master.csv')
df.head(5)
```

```
subset = df[df['generation'].isin(['Silent', 'Boomers', 'Millenials', 'Generation X'])]
sns.set_style('darkgrid')
fig, ax = plt.subplots()
fig.set_size_inches(15,10)
ax = sns.stripplot(x = 'generation', y = 'suicides_no', data = subset, jitter = True, linewidth = 1)
ax.set_title('Generation v/s Suicide')
```

```
subset = df[df['sex'].isin(['male', 'female'])]
sns.set_style('darkgrid')
fig, ax = plt.subplots()
fig.set_size_inches(15,8)
ax = sns.stripplot(x = 'sex', y = 'suicides_no', data = subset, jitter = True, linewidth = 1)
ax.set_title('Gender v/s Suicides')
```

```
X = df[['population', 'GdpPerCapita', 'year']]
Y = df['suicides_no']
```

```
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X,Y)
```

```
import statsmodels.api as sm
X = sm.add_constant(X)
lm_1 = sm.OLS(Y,X).fit()
print(lm_1.summary())
```
